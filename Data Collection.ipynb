```python
healthy_patients=[]
sick_patients=[]

yourpath = 'Dataset/Healthy'
for root, dirs, files in os.walk(yourpath, topdown=False):
    for name in files:
        if( '_f.jpg' in name):
            p_id=name.split('_')[0].split('p')[1]
            healthy_patients.append(int(p_id))

yourpath = 'Dataset/Sick'
for root, dirs, files in os.walk(yourpath, topdown=False):
    for name in files:
        if( '_f.jpg' in name):
            p_id=name.split('_')[0].split('p')[1]
            sick_patients.append(int(p_id))

            
```

```python
def get_data():
    data_dir = 'Test'
   
    transform = transforms.Compose([
        #transforms.RandomRotation(20),
        transforms.Resize((640,640)),
        #transforms.RandomHorizontalFlip(),
        transforms.Grayscale(),
        transforms.ToTensor()])

    train_set = torchvision.datasets.ImageFolder(data_dir , transform=transform)
    test_set = torchvision.datasets.ImageFolder(data_dir , transform=transform)

    train = torch.utils.data.DataLoader(train_set, batch_size=5, shuffle=True)
    test = torch.utils.data.DataLoader(test_set, batch_size=5, shuffle=False)

    return train, test
```

```python
train, test=get_data()
```

```python

def train_imshow():
    classes = ( 'Sick') # Defining the classes we have
    dataiter = iter(test)
    images, labels = dataiter.next()
    
    
    fig, axes = plt.subplots(figsize=(10, 4), ncols=5)
    for i in range(5):
        ax = axes[i]
        ax.imshow(images[i].permute(1, 2, 0)) 
        ax.title.set_text(' '.join('%5s' % classes[labels[i]]))
    plt.show()
    
```
